```python
import json
import os
import matplotlib.pyplot as plt

CHECKPOINTS_PATH = os.path.join("checkpoints")
CHECKPOINT_NAME = 'SA-LSTM_50_epochs_base.json'
```

```python
with open(os.path.join(CHECKPOINTS_PATH, CHECKPOINT_NAME)) as f:
    history = json.load(f)

print(history)
```

```python
train_loss = [tl['total'] for tl in history['train_loss']]
val_loss = [tl['total'] for tl in history['val_loss']]
print(train_loss, len(train_loss))
```

```python
plt.plot(train_loss)
plt.plot(val_loss)
plt.legend(['train', 'val'])
plt.show()
```

```python
for metric in ['Bleu_1', 'Bleu_2', 'Bleu_3', 'Bleu_4', 'ROUGE_L', 'CIDEr']:
    val_score = [ts[metric] for ts in history['val_score']]
    print(metric, val_score)
```
